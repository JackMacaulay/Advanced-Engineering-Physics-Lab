```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress  

plt.rcParams.update({'font.size': 10, 'figure.dpi': 200}) 
```

```python
# Path to the CSV file
file_path = 'Feb27_Testing\Temp_Sweep.csv'
data = pd.read_csv(file_path, skiprows=1)
```

```python
temp = data['Temperature']
resistance = data['Resistance']

minval = np.min(resistance) 

resistance = resistance - minval

start = 280
stop = 350

resistance = resistance[start:stop]
temp = temp[start:stop]

plt.plot(temp,resistance)
```

```python
linearized_x = 1/temp
linearized_y = np.log(resistance*temp**(3/2))
```

```python
# Performing linear regression
slope, intercept, r_value, p_value, std_err = linregress(linearized_x, linearized_y)

plt.plot(linearized_x,linearized_y,'r.')
```

```python
k_B = 1.380649e-23
Eg = 2*slope*k_B/(1.60218e-19)

print(Eg)
```
