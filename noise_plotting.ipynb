```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Specify the folder path where the data files are located
folder_path = r"50_Ohm_Terminator_Sensitivity_10uV"

# Get a list of all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith(".csv")]

# Loop through each CSV file and plot the data
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)

    # Assuming your CSV file has columns 'Frequency (Hz) - Plot 0' and 'X noise (V per root Hz) - Plot 0'
    x_values = df['Frequency (Hz) - Plot 0']
    y_values = df['X noise (V per root Hz) - Plot 0']*(10**9)

    # Plot the data with log scale on both axes, using the file name as a label
    plt.loglog(x_values, y_values, label=f'Data from {file}')

# Add labels and title
plt.xlabel('Frequency (Hz)')
plt.ylabel('X noise (nV per root Hz)')
plt.title('Electrical Noise Measurements')

# Add legend if needed
plt.legend()

# Set log scale for grid lines
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xscale('log')
plt.yscale('log')

# Show the plot
plt.show()
```
